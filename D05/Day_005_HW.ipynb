```python
print(app_train)
```

```python
print('平均數:', app_train['AMT_INCOME_TOTAL'].mean())                        # 計算平均數
print('\n總和:', app_train['AMT_INCOME_TOTAL'].sum())                         # 計算總合
print('\n前1~5筆資料的總和:', app_train['AMT_INCOME_TOTAL'].cumsum().head())   # 計算前N筆data的總和
print('\n一些相關的信息:\n', app_train['AMT_INCOME_TOTAL'].describe())         # 把一些相關的信息都列出來
print('\n總共有幾筆資料:', app_train['AMT_INCOME_TOTAL'].count())              # 計算總共有幾筆資料
print('\n最小值:', app_train['AMT_INCOME_TOTAL'].min())                       # 回傳最小值
print('\n最大值:', app_train['AMT_INCOME_TOTAL'].max())                       # 回傳最大值
print('\n中位數:', app_train['AMT_INCOME_TOTAL'].median())                    # 回傳中位數
print('\n變異數', app_train['AMT_INCOME_TOTAL'].var())                        # 回傳變異數
print('\n標準差', app_train['AMT_INCOME_TOTAL'].std())                        # 回傳標準差
print('\n偏態係數:', app_train['AMT_INCOME_TOTAL'].skew(skipna=True))         # 回傳偏態係數(檢測一組資料是否對稱，如果SK是正的=>偏右，反之。負的=>偏左)
print('\n峰度:', app_train['AMT_INCOME_TOTAL'].kurt())                        # 回傳峰度
```
